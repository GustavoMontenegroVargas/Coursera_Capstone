```python
import pandas as pd
```

```python
C_postalcode = pd.read_html('https://micodigopostal.org/ciudad-de-mexico/cuauhtemoc/')
C = C_postalcode[0]
C.head()
```

```python
C.drop((4),axis=0,inplace=True)

C.reset_index(level=0, inplace=True)

C.drop('index',axis=1,inplace=True)
C.head(10)
```

```python
C = C.rename(columns={'Código Postal' : 'Postal Code'})
C.head()
```

```python
C = C.rename(columns={'Asentamiento▼' : 'Neighborhood'})
C.head()
```

```python
C.drop({'Zona','Mapa'},axis=1,inplace=True)
C.drop('Tipo de Asentamiento',axis=1,inplace=True)
C.head()
```

```python
C_1 = C[['Neighborhood','Postal Code']]
C_1.head(20)
```
