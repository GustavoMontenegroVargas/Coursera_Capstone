```python
import pandas as pd
```

```python
T_postalcode = pd.read_html('https://micodigopostal.org/ciudad-de-mexico/tlahuac/')
T = T_postalcode[0]
T.head()
```

```python
T.drop((4),axis=0,inplace=True)

T.reset_index(level=0, inplace=True)

T.drop('index',axis=1,inplace=True)
T.head(10)
```

```python
T = T.rename(columns={'Código Postal' : 'Postal Code'})
T = T.rename(columns={'Asentamiento▼' : 'Neighborhood'})
T.head()
```

```python
T_1 = T[['Neighborhood','Postal Code']]
T_1.head()
```

```python
T_1.shape
```

```python
!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
```

```python
address = '{}, CDMX'.format(T_1['Postal Code'][0])

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude

print('The coordinates of {} are: Latitude {}, Longitude{}'.format(T_1['Neighborhood'][0], latitude, longitude))
```

```python
Latitude_list=[]
Longitude_list=[]

N = len(T_1)
for i in range(N):
    address = '{}, CDMX'.format(T_1['Postal Code'][i])

    geolocator = Nominatim(user_agent="ny_explorer")
    location = geolocator.geocode(address)
    latitude = location.latitude
    longitude = location.longitude
    Latitude_list.append(latitude)
    Longitude_list.append(longitude)
 
```

```python
   
print(Latitude_list)
print(Longitude_list)
```

```python
print(len(Latitude_list))
print(len(Longitude_list))
```

```python
T_1['Latitude'] = Latitude_list
T_1['Longitude'] = Longitude_list
T_1.head()
```
